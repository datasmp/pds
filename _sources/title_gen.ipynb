```python
import matplotlib.pyplot as plt
import random
```

```python
title_list = ['cover', 'python for data science', 'acknowledgments', 'Regular Expressions', 'Regular Expressions Exercises', 'Files', 
              'Files Exercises', 'Web APIs', 'Web APIs Exercises','Web Crawling', 'Web Crawling Exercises',
              'Series', 'Series Exercises','DataFrames', 
              'DataFrames Attributes and Methods', 'DataFrames Advanced Methods', 'DataFrames Visualization',
              'DataFrames Exercises', 'visualization', 'Visualization Matplotlib', 'Visualization Seaborn', 
              'Visualization Examples', 'numpy arrays', 'Missing Values', 'Statistics', 'datasets', 'Stock Data Analysis']
```

```python
color_set = ['r', 'k', 'g', 'navy', 'orange', 'pink', 'purple']
random.seed(0)

for title in title_list:
    plt.figure(figsize=(12,2))
    plt.xlim(-1,len(title))
    plt.ylim(0,2)
    plt.axis('off')
    for i in range(len(title)):
        
        if random.randint(0,1): j = title[i].upper()
        else: j = title[i]
            
        color_rand = color_set[random.randint(0, len(color_set)-1)]
        size_rand  =  random.randint(40,85)
        rotation_rand = random.randint(0,90)
        plt.text(i, 1, j,{'c':color_rand, 'size':size_rand, 'rotation':rotation_rand})
        plt.savefig('title_pict/'+title.replace(' ', '_')+'.png')
```
