# Visualization

## Matplotlib
- In this section we will use `matplotlib.pyplot`, imported as `plt`.

````python
import matplotlib.pyplot as plt
````

### Scatterplot
- *(x_coordinate, y-coordinate)* should be given for a single point
- *(x_coordinate sequence, y-coordinate sequence)* should be given for multiple points
    - sequence can be a tuple, list, series 

````python
# single point
plt.scatter(5,10);    # x=5, y=10
````

- In the following code the points are: $(2,10), (4,3), (7,6)$.

````python
# multiple points
plt.scatter([2,4,5],[10,3,6]);  
````

### Figure Size

````python
plt.figure(figsize=(10,3))    # width:10, height:3
plt.scatter([2,4,5],[10,3,6]);    
````

### Color

````python
plt.scatter([2,4,5],[10,3,6], c='red');    # instead of 'red' it s enough to write 'r' 
````

### Size

````python
plt.scatter([2,4,5],[10,3,6], s=300);    # size=300
````

### Markers and Scatters
For more marker styles check:
- https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers


````python
# plus symbol
plt.scatter([2,4,5],[10,3,6], marker='+');
````

````python
# square
plt.scatter([2,4,5],[10,3,6], marker='s');
````

### Transparancy

````python
plt.scatter([2,4,5],[10,3,6], alpha=0.3);
````

### Linewidth and Scatters

````python
plt.scatter([2,4,5],[10,3,6], c='r', s=200, linewidths=5, edgecolor='b');
````

### Title

````python
plt.title('Three Square Points')
plt.scatter([2,4,5],[10,3,6], c='r', s=200, marker='s');
````

````python
plt.title('Three Points', fontsize=20, loc='right', pad=100, c='b')
plt.scatter([2,4,5],[10,3,6], c='r', s=200, marker='s');
````

### Suptitle

````python
plt.title('Three Points', fontsize=20, loc='right', pad=100, c='b')
plt.suptitle('Different Colors', fontsize=15, c='g', y=1)
plt.scatter([2,4,5],[10,3,6], c='r', s=200, marker='s');
````

### Grid

````python
plt.scatter([2,4,5],[10,3,6])
plt.grid();
````

````python
# only vertical
plt.scatter([2,4,5],[10,3,6])
plt.grid(axis='x');
````

````python
# only horizontal
plt.scatter([2,4,5],[10,3,6])
plt.grid(axis='y');
````

### Legend

````python
plt.scatter([2,4,5],[10,3,6], c='r', label='reds')
plt.scatter([5,1,9],[4,8,7],  c='g', label='greens')
plt.legend();
````

````python
# location of legend: lower left

plt.scatter([2,4,5],[10,3,6], c='r', label='reds')
plt.scatter([5,1,9],[4,8,7],  c='g', label='greens')
plt.legend(loc='lower left');
````

### Axis Labels

````python
plt.scatter([2,4,5],[10,3,6])
plt.xlabel('x_coordiante')
plt.ylabel('y_coordiante');
````

### Axis Ticks

````python
plt.scatter([2,4,5],[10,3,6])
plt.xticks([2,4,5], ['Washington', 'Alabama','Virginia']);
````

````python
# rotation
plt.scatter([2,4,5],[10,3,6])
plt.xticks([2,4,5], ['Washington', 'Alabama','Virginia'], rotation=90);
````

### Text

````python
plt.scatter([2,4,5],[10,3,6])
plt.text(4.05,3,'Alabama');   # starts from the point (4.05,3)
````

### Facecolor

````python
plt.figure(facecolor='g')
plt.scatter([2,4,5],[10,3,6]);
````

````python
plt.axes().set_facecolor('yellow')
plt.scatter([2,4,5],[10,3,6]);
````

### Subplot

````python
# number of rows: 1
# number of columns: 3

plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
plt.scatter([2,4],[8,7],c='r')
plt.subplot(1,3,2)
plt.scatter([1,2],[5,12],c='b')
plt.subplot(1,3,3)
plt.scatter([6,10],[1,2],c='g');
````

### Lineplot

````python
plt.plot([1,3],[7,10]);  # start from (1,3) up to (3,10)
````

### Linewidth and Lines


````python
plt.plot([1,3],[7,10], linewidth=7);
````

### Color and Lines

````python
plt.plot([1,3],[7,10], c='r');
````

### Linestyle and Lines

````python
plt.plot([1,7],[3,10], c='r', linestyle='dashed')
plt.plot([0,8],[12,3], c='g', linestyle='dotted');
````

### Markers and Lines

````python
plt.plot([1,3],[7,10], c='r', marker='*', markersize=10);
````

````python
# more points
plt.plot([1,3,5,7,9,11,13,13],[7,10,1,8,3,6,2,1], c='r', marker='*', markersize=20);
````

### Pie Chart

````python
expenses = ['housing', 'tuition', 'transportation', 'supplies', 'food']
cost = [700, 1000, 200, 100, 500 ]
color_set = ['y', 'r', 'g', 'orange', 'navy']
````

````python
plt.pie(cost, colors = color_set, labels=expenses, radius=1, autopct='%.0f');
````

### Bar Graph

````python
plt.bar(expenses, cost);
````

````python
# horizontal
plt.barh(expenses, cost, color='r' );
````

### Boxplot

````python
plt.boxplot([8,7,3,2,1,5,6,7,8]);
````

## Seaborn

````python
import seaborn as sns
````

### Scatterplot

````python
sns.scatterplot(x=[2,4,7],y=[10,4,6], color='r', s=100);
````

### Lineplot

````python
sns.lineplot(x=[2,4,7],y=[10,4,6], color='r');
````

### Barplot

````python
sns.barplot(x=cost, y=expenses);
````

### Box Plot

````python
sns.boxplot([8,7,3,2,1,5,6,7,8]);
````

### Displot

````python
sns.displot([8,7,3,2,1,5,6,7,8,1,3,2,5]);
````

### Histplot

````python
sns.histplot([8,7,3,2,1,5,6,7,8,1,3,2,5]);
````

### Violinplot

````python
sns.violinplot([8,7,3,2,1,5,6,7,8,1,3,2,5]);
````

### Violinplot

````python
sns.swarmplot([8,7,3,2,1,5,6,7,8,1,3,3,3,3,3,3,3,3,3,2,5,5,5,5,5,5,5,5]);
````

````python
sns.swarmplot([8,7,3,2,1,5,6,7,8,1,3,3,3,3,3,3,3,3,3,2,5,5,5,5,5,5,5,5], c='yellow');
sns.violinplot([8,7,3,2,1,5,6,7,8,1,3,3,3,3,3,3,3,3,3,2,5,5,5,5,5,5,5,5], color='red');
````
