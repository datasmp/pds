# Dataframe
A DataFrame is a two-dimensional data structure characterized by labeled axes (rows and columns). 
- Resembling a Microsoft Excel worksheet, it accommodates mixed data types efficiently.
- Primarily utilized for data wrangling and manipulation.
- DataFrames offer numerous methods for tasks such as data cleaning, exploration, and preprocessing.
- They support vectorized operations across rows or columns and facilitate visualization.
- The main data structure of the pandas library, the DataFrame, requires importing the library to create and manipulate dataframes and utilize its methods. 
- It's commonplace to abbreviate pandas as pd for brevity in code.
    - `import pandas as pd`

### Slicing
Slicing can be done either by using the row and column indexes or labels.
1. Row and column indexes: Use `iloc[row index, column index]`
2. Row and column labels : Use ` loc[row label, column label]`

Let's remember the small dataframe *df_sh*.

````python
df_sh
````

- Let's try to access the height of Tim which is 140 by using *loc* and *loc*.
    - row label of 140 is *Tim*.
    - column label of 140 is *Height*.

````python
df_sh.loc['Tim', 'Height']
````

- Let's try to access the height of Tim which is 140 by usin *iloc*.
    - row index of 140 is index of *Tim* in ['John', 'Liz', 'Tim']  which is 2.
    - column index of 140 is index of *Height* in ['State', 'Height']  which is 1.

````python
df_sh.iloc[2,1]
````

- You can also do slicing as in strings, tuples, and lists.
- Slicing with *loc* (labels) has a small difference.
    - *loc* is inclusive for the end label which means the end label is included.
- Example: The column labels of *df_sh* dataframe ['State', 'Height']
    - *df_sh.loc['Liz':,:'Height']* inludes he columns up to and including 'Height'.

````python
# rows starting from index 1 (includes 1): row indexes: 1, 2
# columns up to index 1 (does not include 1): column index: 0

df_sh.iloc[1:,:1]
````

````python
# rows starting from  'Liz' (includes 'Liz'): row labels: 1, 2
# columns up to label 'Height' (includes 'Height'): column label: 'State', 'Height'

df_sh.loc['Liz':,:'Height']
````

### Filtering
The boolean expressions can be used to select particular parts of the dataframe under some conditions.
- For example *df['column_label'] > 90* returns a series which has a boolen value for each row depending on whether the condtion is True or False for that row.

````python
# returns a boolean value for each row.
df_grades['Test-3']>90
````

- Two conditions combined with a boolean operator.

````python
(df_grades['Test-2']>65) & (df_grades['Test-3']>90)
````

- These boolean values can be used in a square bracket to select a part of the dataframe.

**Example:** The following code selects the rows for which *Test-3* is grade is more than 90.

````python
df_grades[ df_grades['Test-3']>90 ]
````


**Example:** The following code selects the rows for which *Test-2* is grade is more than 65 and the *Test-3* is grade is more than 90.

````python
df_grades[ (df_grades['Test-2']>65) & (df_grades['Test-3']>90) ]
````

**Example:** The following code selects the rows for which *Gender* is male and the *Final* is grade is more than 95.

````python
df_grades[ (df_grades['Gender']=='M') & (df_grades['Final']>95) ]
````

## Iterations and Dataframes
Dataframes can be iterated through their column and row labels or indexes.
- You can use
    - *df.index* for row labels
    - *df.columns* for column labels
    - *range(len(df))* or *range(df.shape[0])* for row indexes
    - *range(df.shape[1])* for column indexes

````python
df_sh
````

````python
# row labels
for row in df_sh.index:
    print(row)
````

````python
# column labels
for col in df_sh.columns:
    print(col)
````

````python
# access al values using row and column labels
for row in df_sh.index:
    for col in df_sh.columns:
        print(df_sh.loc[row, col])
````

````python
for i in range(df_sh.shape[0]):    # you can also use len(df_sh)
    print(i)
````

````python
# column indexes
for j in range(df_sh.shape[1]):
    print(j)
````

````python
# access al values using row and column indexes
for i in range(df_sh.shape[0]):
    for j in range(df_sh.shape[1]):
        print(df_sh.iloc[i,j])
````
