## Question-1
Find all unique website URLs on the http://quotes.toscrape.com/ website using regular expressions.

- Hint: The URLs are located within `<div>` HTML elements and start with http.

```python
import requests
from bs4 import BeautifulSoup

url = 'http://quotes.toscrape.com/'
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')
```

```python
div_list = soup.find_all('div')
div_list
```

```python
import re
http_list = []
for div_item in div_list:
    http_list += re.findall('(http[^"]+)"' , div_item.decode())
set(http_list)
```

## Question-2
Find all unique font sizes on the http://quotes.toscrape.com/ website using regular expressions.

- Hint: The font are located within `<a>` HTML elements.

```python
import requests
from bs4 import BeautifulSoup

url = 'http://quotes.toscrape.com/'
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

tags = soup.find_all('a', class_='tag')
tags
```

```python
tags[-2]
```

```python
import re
fontsize_list = []
for tag in tags:
    fontsize_list += re.findall('font-size: ([\d]+)' , tag.decode())
set(fontsize_list)
```

## Question-3
Find all the headlines and their corresponding URLs on the Hacker News website using regular expressions.

- Hint: The headlines are contained within <a> HTML elements.
- Ensure that you extract 30 headlines along with their URLs.

```python
import requests
from bs4 import BeautifulSoup

url = 'https://news.ycombinator.com/'
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

soup
```

```python
a_list = soup.find_all('a')
a_list
```

```python
import re
url_list, headlines_list = [], []
for a_item in a_list:
    headline = re.findall('https.+">([^<]+)' , a_item.decode())
    if len(headline) != 0:
        url_list += re.findall('(https[^"]+)' , a_item.decode())    
        headlines_list += headline
headlines_list = headlines_list[:-3]
url_list = url_list[:-3]
```

## Question-4
Find all book titles on the https://books.toscrape.com/ website using regular expressions.

- Consider using find_all() instead of findall() at some point.

```python
import requests
from bs4 import BeautifulSoup

url = 'https://books.toscrape.com/'
response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

soup
```

```python
a_list = soup.find_all('a')
a_list
```

```python
import re
title_list = []
for a_item in a_list:
    title_list += re.findall('img alt="([^"]+)' , a_item.decode())
title_list
```
